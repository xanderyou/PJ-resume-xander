```python
import requests
from bs4 import BeautifulSoup
import re


new_list = []
url = 'https://money.udn.com/money/index'

# 建立headers
headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:121.0) Gecko/20100101 Firefox/121.0'}

#取得網頁內容.get
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
# print(response.text)

# 抓取即時區域所有新聞title以及新聞內容連結link，並將每篇新聞的連結link append到new_list內
li_tags=soup.find('li', {'class':'latest story'})
# print(li_tags)
soup=BeautifulSoup(str(li_tags), 'html.parser')
li_tags=soup.find_all('div', {'class':'story__content'})
# print(li_tags)
soup2=BeautifulSoup(str(li_tags), 'html.parser')
a_tags=soup2.find_all("a", {"data-slotname":"list_最新"})
# print(a_tags)

# 提取標題與網頁連結
for i in a_tags:
    title=i['title'].replace(" ","").replace("\n","")
    if len(title)>0:
        dw_link="https://money.udn.com"+i["href"].replace(" ","").replace("\n","")
        print('title: {}, link: {}'.format(title, dw_link))
        new_list.append(dw_link)
    
# print(new_list)
```

```python
from time import sleep
import os
from bs4 import BeautifulSoup
from urllib.request import urlretrieve

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)

        
def save_news(url):
    # headers
    headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:121.0) Gecko/20100101 Firefox/121.0'}
    response = requests.get(url, headers=headers)
    sleep(1)
    
    soup = BeautifulSoup(response.text, 'html.parser')
    # 使用find_all抓取每篇新聞的title，產生之list放到new_title變數內
    new_title=soup.find_all('h1', {'id':'story_art_title'})

    save_path = './news/'+new_title[0].text.replace(':','').replace(" ","").replace("\n","")
    
    article_body = soup.find_all('section',{'id':'article_body'})
    print('new title: {}'.format(new_title[0].text))

    # 創建資料夾
    createFolder(save_path)
    soup2 = BeautifulSoup(str(article_body[0]), 'html.parser')


    # 儲存文章，抓取文章內p tags，並將新聞內文寫入contents.txt檔案內
    with open(save_path+'/contents.txt', 'w', encoding='utf-8') as f:  
        print('new content: ')
        p_tags=soup2.find_all('p')
        for i in p_tags:
            print(i.text)
            f.write(i.text)

    ## 儲存圖片，抓取文章內所有img tags，並將圖片下載並存入對應新聞資料夾下  
    # print('save image: ')
    img_tags=soup2.find_all('img')
    for i in img_tags:
        # print(i['src'])
        img_path=i['src']
        img_path=img_path.replace(" ","")  
        img_name=i['title'].replace(" ","")
        urlretrieve(img_path, save_path+'/'+i['title']+'.jpg')
    print('==========\n\n')
    


    
for new_url in new_list:
    print(new_url)
    save_news(new_url)
```
